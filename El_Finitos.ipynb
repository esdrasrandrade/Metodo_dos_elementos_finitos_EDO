```python
import numpy as np
from scipy.integrate import quad
```

```python
#Matriz de Rigidez
dominio = 1.0 #comprimento(b-a)
ne = 4 #Número de elementos
nn = ne + 1 #Número de nós
h = dominio/ne
```

```python
def L1(x):
  return 1 - h/x
```

```python
def L2(x):
  return x/h
```

```python
def dL1(x):
  return -1/h
```

```python
def dL2(x):
  return 1/h
```

```python
L = [L1, L2]
dL = [dL1, dL2]

def integrando(x,i,j):
  return -dL[i](x) * dL[j](x) + L[i](x) * L[j](x)

Ke = np.zeros((2,2))
for i in range(2):
  for j in range (2):
    Ke[i,j], erro = quad(integrando, 0, dominio, args=(i,j))
print("Matriz de rigidez do elemento: ")
print(Ke)
K = np.zeros((nn,nn))
for e in range (ne):
  nos = [e, e+1]
  for i in range(2):
    K[nos[i], nos[j]] += Ke[i,j]

print("\nMatriz de rigidez global: ")
print(K)
```
